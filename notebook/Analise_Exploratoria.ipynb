```python
# Importação do Pandas
import pandas as pd
```

```python
    # File Path
source = '../data/electronics_sales_raw.csv'
```

```python
# Leitura do DataFrame
df = pd.read_csv(source)
```

```python
# Visualizando o DataFrame
df.head(10)
```

```python
# Verificando informações sobre o DataFrame
df.info()
```

```python
# Descrevendo o DataFrame
df.describe()
```

```python
# Verificando quantidade de NA
df.isna().sum()
```

```python
# Transforma os nomes das colunas em uma lista de Python tradicional
lista_de_colunas = df.columns.tolist()
print(lista_de_colunas)
```

```python
# Removendo Colunas de Date, Numerais, Nomes e ID
df_sem_numeros = df.select_dtypes(exclude=['number'])
df_sem_numeros = df_sem_numeros.drop(columns=['first_purchase_date', 'last_purchase_date','order_date','customer_id','customer_name','product_id'])
```

```python
df_sem_numeros.head()
```

```python
# Verificando Grupos Únicos
for coluna in df_sem_numeros.columns:
    categorias_unicas = df[coluna].unique()

    print(f"📌 Coluna: '{coluna}'")
    print(f"Valores únicos: {categorias_unicas}")
```

```python
colunas_texto = df.select_dtypes(include=['object']).columns

# For loop
for coluna in colunas_texto:

    # Coluna Original
    coluna_original = df[coluna]

    # Criada a versão limpa
    coluna_limpa = coluna_original.str.strip()

    # Comparamos as duas:
    # Uso do dropna() para garantir que valores vazios (NaN) não sejam contados por engano.
    qtd_modificados = (coluna_original.dropna() != coluna_limpa.dropna()).sum()

    # Mostrando resultado
    if qtd_modificados > 0:
        print(f" Coluna '{coluna}': {qtd_modificados} itens tinham espaços e foram corrigidos.")
    else:
        print(f" Coluna '{coluna}': Já estava limpa (0 modificações).")

    # Substituindo a coluna no DataFrame principal
    df[coluna] = coluna_limpa
```

```python
df.head()
```

```python
# Verficando Ordens de Compra Duplicadas
total = df.duplicated(subset=['order_id']).sum()
print(f"Total de duplicados: {total}")
```

```python
df_com_numeros = df.select_dtypes(exclude=['object'])
df_com_numeros.head()
```

```python
df_tratado = df_com_numeros[['quantity','unit_price','discount_pct','operating_expenses','cash_balance','debt_balance',
                             'monthly_burn','churn_flag']]
df_tratado.head(10)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Distribuição das Colunas Int e Float
fig, axes = plt.subplots(nrows=2, ncols=4, figsize=(20, 8))
axes = axes.flatten()

colunas = ['quantity','unit_price','discount_pct','operating_expenses','cash_balance','debt_balance',
                             'monthly_burn','churn_flag']

for i, col in enumerate(colunas):
    sns.histplot(data=df_tratado, x=col, kde=True, ax=axes[i], color='steelblue')
    axes[i].set_title(col)
    axes[i].set_xlabel('')

plt.suptitle('Distribuição das Colunas Numéricas', fontsize=14, y=1.02)
plt.tight_layout()
plt.show()
```
